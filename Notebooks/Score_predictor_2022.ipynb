```python
df['current_score'] = df.groupby('match_id')['total_runs'].cumsum()
```

```python
df.head(10)
```

```python
df['balls_bowled'] = (df['over'].astype(int) * 6) + df['ball'].astype(int)
df.head(10)
```

```python
df['balls_left'] = 120 - df['balls_bowled']
df['balls_left'] = df['balls_left'].apply(lambda x:0 if x<0 else x)
df.sample(5)
```

```python
df['player_dismissed'] = df['player_dismissed'].apply(lambda x: 0 if x == 0 else 1)
```

```python
df['player_dismissed'].dtype
```

```python
df
```

```python
df['player_dismissed'] = df.groupby('match_id')['player_dismissed'].cumsum()
df['wickets_left'] = 10 - df['player_dismissed']
```

```python
df.tail(30)
```

```python
df['crr'] = (df['current_score']*6)/df['balls_bowled']
```

```python
df
```

```python
# groups = df.groupby('match_id')

# match_ids = df['match_id'].unique()
# last_five = []
# for id in match_ids:
#     last_five.extend(groups.get_group(id).rolling(window=30).sum()['total_runs'].values.tolist())
```

```python
groups = df.groupby('match_id')
match_ids = df['match_id'].unique()
last_three_run = []
for id in match_ids:
    # Select only the 'runs' column, then apply rolling and sum
    runs_sum = groups.get_group(id)['total_runs'].rolling(window=18).sum()
    last_three_run.extend(runs_sum.values.tolist())
```

```python
df['last_three_runs'] = last_three_run
```

```python
df.tail(30)
```

## Train-Test Split

```python
X = final_df.drop(columns=['total_runs_x'])
y = final_df['total_runs_x']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score,mean_absolute_error
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import SelectKBest,chi2
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV
```

```python
trf1 = ColumnTransformer(
    transformers=[
        ('trf', OneHotEncoder(sparse_output=False, drop='first'), ['batting_team', 'bowling_team', 'city'])
    ],
    remainder='passthrough'
)
```

## XGBRegressor

```python
pipe_xg = Pipeline(steps=[
    ('step1',trf1),
    ('step2',StandardScaler()),
    ('step3',XGBRegressor(n_estimators=1200,learning_rate=0.1,max_depth=12,random_state=1))
])
```

```python
pipe_xg.fit(X_train,y_train)
```

```python
y_pred = pipe_xg.predict(X_test)
```

```python
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```
